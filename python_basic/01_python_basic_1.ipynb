## ● 변수와 타입

파이썬에서는 변수를 사용하기 전에 무언가를 기술할 필요가 없습니다.  
다음과 같이 값을 대입하는 것부터 기술할 수 있습니다.  

````python
a = 123
````

파이썬에서는 변수에 대해 타입을 명시할 필요가 없습니다.  
예를 들어 정수 타입 변수에 문자열을 대입하면, 그 변수는 문자열 타입의 변수가 됩니다.  
파이썬의 주요한 타입은 다음과 같습니다.

````python
a = 123             # 정수 타입(int)
b = 123.456         # 부동 소수점 타입(float)
c = "Hello World!"  # 문자열 타입(str)
d = True            # 부울 타입(bool)
e = [1, 2, 3]       # 리스트 타입(list)
````

\#는 주석을 나타내고, 같은 행에서 해당 기호 이후의 내용은 코드로 인식되지 않습니다.  

그리고 bool 타입의 값은 수치로 다룰 수 있습니다.  
`True`는 1, `False`는 0으로 다루어집니다.  
다음 예에서는 `True`와 `False`를 더하고 있습니다. 결과는 0과 1의 합인 1이 됩니다.

````python
a = True; b = False
print(a+b)
````

파이썬에서는 위와 같이 `;`(세미콜론)으로 구분해 한 행에서 여러 처리를 작성할 수 있습니다.

그리고 부동 소수점 타입의 값은 지수 표기를 사용할 수 있습니다. 다음과 같이 `e`를 사용해 소수를 표기할 수 있습니다.

````python
1.2e5   # 1.2 x 10의 5제곱 120000
1.2e-5  # 1.2 x 10의 -5제곱 0.000012
````

## ● 리스트

리스트는 여러 값을 모아서 취급할 때 사용합니다.  
리스트는 전체를 `[]`로 감싸고, 각 요소는 `,`로 구분합니다.  
파이썬의 리스트는 모든 타입의 값을 저장할 수 이;ㅆ으며, 리스트 안에 리스트를 저장할 수도 있습니다.  
리스트의 각 요소에 접근할 때는 인덱스를 사용하며, 요소를 추가하거나 교체할 수도 있습니다.

````python
a = [1, 2, 3, 4, 5]     # 리스트 생성

b = a[2]                # 3번째 요소를 얻는다
print(b)

a.append(6)             # 마지막에 요소를 추가한다
print(a)

a[2] = 7                # 요소를 교체한다
print(a)
````

## ● 튜플

튜플은 리스트와 마찬가지로 여러 값을 모아서 다룰 때 사용합니다. 하지만 요소 추가, 삭제, 교체 등은 할 수 없습니다.  
튜플은 전체를 `()`로 감싸고, 각 요소는 `,`로 구분합니다.  
요소를 변경할 예정이 없을 때는 리스트 보다 튜플을 사용하는 편이 좋습니다.

````python
a = (1, 2, 3, 4, 5)     # 튜플 생성

b = a[2]                # 3번째 요소를 얻는다
print(b)
````

요소가 하나 뿐인 튜플을 생성할 때는 다음과 같이 요소 바로 뒤에 `,`를 입력해야 합니다.

````python
(3,)
````

리스트나 튜플의 요소는 다음과 같이 모아서 변수에 대입할 수 있습니다.

````python
a = [1, 2, 3]
a1, a2, a3 = a
print(a1, a2, a3)

b = (4, 5, 6)
b1, b2, b3 = b
print(b1, b2, b3)
````

## ● 딕셔너리

딕셔너리는 키와 값의 조합으로 데이터를 저장합니다.  
다음은 파이썬 딕셔너리의 사용 예입니다.  
문자열을 키로 하는 딕셔너리를 생성하고 값을 얻거나 교체하고, 요소를 추가합니다.


````python
a = {"Artificial":1, "Intelligence":2}  # 딕셔너리 생성
print(a["Artificial"])   # "Apple" 키를 가진 값을 얻는다

a["Intelligence"] = 7  # 요소 교체
print(a["Intelligence"])

a["ML"] = 3  # 요소 추가
print(a)
````

## ●セット  
セットはリストと似ていますが、重複した値の要素をもつことができません。  
タプルと異なり、要素の追加や削除が可能です。

````python
a = [1, 1, 2, 3, 4, 4, 5, 5, 5]  # リスト
print(a)

b = set(a)  # セットに変換
print(b)

b.add(6)  # 値を追加
print(b)

b.remove(3)  # 値を削除
print(b)
````

## ● if문

분기에는 `if`문을 사용합니다.  
`if` 조건을 만족하지 않으면 `elif` 조건을 위에서부터 순서대로 판정합니다.  
이 조건들을 모두 만족하지 않으면 `else` 안의 처리를 실행합니다.

많은 프로그래밍 언어에서는 분기나 함수 블록을 나타내기 위해 `{ }`를 사용하지만,  
파이썬에서는 블록의 범위를 행의 들여쓰기로 나타냅니다.  
즉 들여쓰기가 없는 행이 나타나면 그 직전의 블록은 종료됩니다.  
들여쓰기를 할 때는 스페이스 4칸을 일반적으로 사용합니다.


````python
a = 7
if a < 12:
    print("Good morning!")
elif a < 17:
    print("Good afternoon!")
elif a < 21:
    print("Good evening!")
else:
    print("Good night!")
````

## ● for문

지정한 횟수만큼 루프를 돌 때는 `for`문을 사용합니다.  
루프를 도는 범위를 지정할 때는 리스트나 `range`를 `in` 연선자와 함께 사용합니다.

`range` 사용 방법은 다음과 같습니다. `[ ]`로 감싼 인수는 생략할 수 있습니다.

```
range([시작 번호,] 종료 번호[, 단계 수])
```

예를 들어 `range(3)`은 0부터 2까지의 범위가 됩니다.

````python
for a in [4, 7, 10]:    # 리스트를 사용한 루프
    print(a)

for a in range(3):      # range를 사용한 루프
    print(a)
````

## ● while문

어떤 조건을 만족하는 동안 루프를 돌 때는 `while`을 사용합니다.

````python
a = 0
while a < 3:  # a가 3보다 작은 동안 루프
    print(a)
    a += 1
````

## ● 컴프리헨션 표기

컴프리헨션은 리스트의 요소를 조작하면서 새로운 리스트를 생성하기 위한 표기법입니다.  
일반적으로 그런 처리는 `for`나 `while`에 의한 루프를 사용하지만, 컴프리헨션을 사용하면 간결하게 코드를 기술할 수 있습니다.  
컴프리헨션 표기는 다음 형식으로 기술합니다.

```
새로운 리스트 = [ 요소에 대한 처리 for 요소 in 리스트 ]
```

리스트 안의 요소를 하나씩 꺼내 요소에 대한 처리를 실행한 상태에서 새로운 리스트를 생성합니다.


````python
a = [1, 2, 3, 4, 5, 6]
b = [c*3+1 for c in a]    # a의 요소를 3배한 뒤 1을 더해 새로운 리스트를 만든다
print(b)
````
